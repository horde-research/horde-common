```python
examples = """
Едыбысына қатысты қатарды белгілеңіз.
A: Қысаң, еріндік, жіңішке.
B: Қысаң, езулік, жіңіщке.
C: Ашық, езулік, жуан.
D: Ашық, езулік, жіңішке.
E: Ашық, еріндік, жіңішке.
Answer: A
**************************************************
С. Мұқанов мұғалімдік оқуды қай қалада оқығанын белгілеңіз.
A: Орынбор.
B: Семей.
C: Омбы.
D: Уфа.
E: Қарағанды.
Answer: C
**************************************************
Бастауыш пен баяндауыштың арасына қойылатын сызықшаны көрсетіңіз.
A: Ағалы-інілі екеуіне сүйсіндік.
B: Амангелді сынды батырын ел ұмытпас.
C: Он-он бесі келді.
D: Көрмес-түйені де көрмес.
E: Үгіт-насихат жұмысын жүргіз.
Answer: D
**************************************************
С. Мұқановтың «Аққан жұлдыз» романының кейіпкер(-лер)і
A: Ермек
B: Алтай
C: Айжан
D: Байбосын
E: Нарқыз
F: Рәзия
G: Шоқан
H: Қайсар
Answer: A
**************************************************
Өздік есімдігі қатысқан тіркесті табыңыз.
A: Сөзшең адам.
B: Өз жұмысы.
C: Біткен іс.
D: Өнерпаз бала.
E: Қайратты жігіт.
Answer: B
**************************************************
Демографиялық саясатта туу санын шектеуді ұстанатын елдер
A: Болгария, Венгрия
B: Қытай, Ресей
C: Ресей, Германия
D: Қытай, Үндістан
E: Франция, Румыния
Answer: A
**************************************************
Бүкіл жер шары беті көлемі (млн. км2)
A: 482
B: 640
C: 750
D: 510
E: 420
Answer: D
**************************************************
Латын Америкасындағы ОПЕК-ке мүше ел:
A: Аргентина.
B: Перу.
C: Бразилия.
D: Венесуэла.
E: Колумбия.
Answer: D
**************************************************
Жердің ортасынан экваторға дейінгі ара қашықтық:
A: 6370 км.
B: 6378 км.
C: 6359 км.
D: 6356 км.
E: 6375 км.
Answer: B
**************************************************
Африка жағалауындағы ең ірі арал
A: Шри-Ланка
B: Сахалин
C: Мадагаскар
D: Канар
E: Крит
Answer: A
**************************************************
Христофор Колумбтың Американы ашқан жылы.
A: 1465ж.
B: 1502ж.
C: 1495ж.
D: 1492ж.
E: 1490ж.
Answer: D
**************************************************
Ұлттық қайта жаңару үкіметін зұлымдық қастандықтардан қорғау туралы заң 21 наурыз 1933 жыл… Кімде-кім қасақана империяның немесе жекелеген… жерінің игілігіне немесе империя жеке жерлер үкіметінің, үкімет партияларының не жеке бірлестіктерінің беделіне зор нұсқан келтіретін, шындыққа сай келмейтін, не шындықты бұрмалайтын пікірлерді айтқаны не таратқаны үшін, егер осы қылмысы үшін одан да ауыр жаза тиісті болмаса, екі жылға дейін қамау жазасына кесіледі. Егер пікір көпшілік алдында айтылып не таратылса, онда үш айға дейін қамауға алынады. Егер қылмысты әрекет империяға зор зиян келтірсе, онда айыпкер каторгаға айдалу жазасына кесіледі.Мәтін бойынша осы заң қабылдаған мемлекетті анықтаңыз:
A: Польша
B: Испания
C: Англия
D: Германия
E: Франция
Answer: E
**************************************************
Британияның бірінші доминионы.
A: Австралия.
B: Пәкістан.
C: Жаңа Зеландия.
D: Үндістан.
E: Канада.
Answer: E
**************************************************
Цинь Шихуанды қызықтырған жерлер:
A: Оңтүстік Қытай теңізінің жағалауындағы тайпалар жері.
B: Көшпелі ғұн тайпалары жері.
C: Сары теңіз жағалауындағы жерлер.
D: Сақтар жері.
E: Шығыс Қытай теңізі жағалауындағы елдердің жері.
Answer: A
**************************************************
Садақ пен жебенің жасалынған кезеңі:
A: Ерте палеолит
B: Мезолит
C: Энеолит
D: Қола дәуірі
E: Неолит
Answer: E
**************************************************
«Өзбек ұлысы», «Өзбек хандығы», «Шайбани ұлысы» сияқты атаулар осы хандыққа тән:
A: Моғол хандығы
B: Әбілқайыр хандығы
C: Ақ Орда
D: Алтын Орда
E: Батыс Сібір
Answer: B
**************************************************
Сарматтарда шыныдан жасалған ыдыстарды
A: Өздері жасады
B: Шығыс елдерден алғызды
C: Пайдаланбады
D: Әйелдері жасады
E: Арнайы шеберлер жасады
Answer: B
**************************************************
Мустьер дәуіріндегі адам ағаш бұтағын бір-біріне үйкей отырып немесе шақпақ тастарды ұру арқылы:
A: Қару жасады.
B: От алды.
C: Тамақ пісірді.
D: Су алды.
E: Еңбек құралын жасады.
Answer: B
**************************************************
Қырық жылдан астам Қазақстан өкіметінің, республика партия ұйымының басшылығында болған
A: Л.Брежнев
B: Д.Қонаев
C: Н.Оңдасынов
D: Ж.Шаяхметов
E: Н.Назарбаев
Answer: B
**************************************************
ХХ ғ. 60 жылдары Н.С.Хрущев партиялық ұйымдарды қандай белгілері бойынша бөлді?
A: Білімі бойынша.
B: Әкімшілік белгілері бойынша.
C: Өндірістік.
D: Аумақтық.
E: Коммунистердің саны бойынша.
Answer: C
**************************************************
Орфографияға назар аудара отырып, ағылшын сөзін табыңыз.География
A: geogrypha.
B: geograuph.
C: geograpy.
D: geogrphya.
E: geography.
Answer: E
**************************************************
Ерекшеленген әрпі басқаша оқылатын сөзді таңдаңыз.
A: Garden
B: Gaze
C: Gentle
D: Gate
E: Geese
Answer: C
**************************************************
Ш.Уәлихановтың еңбегін табыңыз.
A: "Жеті жарғы".
B: "Сот реформасы" туралы.
C: "Жаңа низам".
D: "Уақытша ереже".
E: "Басқару ережесі".
Answer: D
**************************************************
Тиісті есімдікті қойыңыз:І’m sorry І forgot to phone you. -... all rіght.
A: Thіs was
B: That іs
C: Thіs іs
D: Those іs
E: These іs
Answer: B
**************************************************
Ішекқуыстылардың атпа жасушалары тән тек:
A: Теңіз қызыл маржандарына
B: Адамға қауіпті медузаның кейбір түріне
C: Актинияға
D: Гидраға
E: Барлық ішекқуыстыларға
Answer: E
**************************************************
Жас өркен өсіп шығатын бүршік:
A: Гүл бүршігі.
B: Жанама бүршік.
C: Төбе бүршігі.
D: Жапырақты бүршік.
E: Гүлшанақ.
Answer: B
**************************************************
Бауырсорғышта кірпікше:
A: Болмайды.
B: Дернәсілінде болмайды.
C: Дернәсілінде болады.
D: Ересек ағзада болады.
E: Ересек ағзада болмайды.
Answer: C
**************************************************
Құстың қауырсыны дегеніміз:
A: Жіңішке мұртшалар жиынтығы.
B: Ортасында мүйізденген кеуекті өзегі бар түтік.
C: Терінің үстіңгі қабатындағы теріден пайда болған түзіліс.
D: Сояудың екі жағындағы тармақ.
E: Теріге қадалған ұшы.
Answer: C
**************************************************
Хроматофоры тостаған пішінді балдыр:
A: Хламидомонада.
B: Ламинария.
C: Улотрикс.
D: Хлорелла.
E: Спирогира.
Answer: A
**************************************************
«Заңның билігі жоқ жерде, ешқандай мемлекеттік құрылыстың формасына орын жоқ» — деп айтқан философ:
A: Платон
B: Сократ
C: Демокрит
D: Аристотель
Answer: D
**************************************************
Билікке қатынасу дәрежесіне, билікке деген өкілеттілігінің ауқымына байланысты элитаға жатпайтын топ:
A: Төменгі
B: Маргиналдық
C: Ортаңғы
D: Жоғарғы
Answer: B
**************************************************
Сот кәмелеттік жасқа толмаған екі балаға ата-ана табысынан немесе өзге де табыс көздерінен өндіріп алынатын мөлшері:
A: Үштен бір бөлігін
B: Бестен бір бөлігін
C: Алтыдан бір бөлігін
D: Төрттен бір бөлігін
Answer: A
**************************************************
Банктердің шаруашылық жүргізуші субьектілерге (несие алушыларға ақшалай қарыздар беретін несие формасы:
A: Коммерциялық несие
B: Банктік несие
C: Мемлекеттік несие
D: Тұтынушылық несие
Answer: B
**************************************************
Мүлік иесі мен сол мүлікті жалға алған еңбек ұжымы арасындағы шарттың негізінде құрылған кәсіпорын:
A: халықаралық
B: ұлттық
C: Жалға алу
D: мемлекеттік
Answer: C
**************************************************"""

GENERATE_PROMPT = ("""
<system>
You are an expert item-writer specialising in medium-to-hard multiple-choice questions (MCQs) for Kazakh-language cultural knowledge.  
All questions, answer options, correct-answer keys, and rationales **must be written in Kazakh**, while every instruction you read here is in English.
</system>

<findings>
• Human-curated examples show tight stems, homogeneous distractors that differ by only one feature, and even answer-key distribution.  
• Numeric questions cluster distractors within ±5–10 % of the true value.  
• Tags are useful for probing specific reasoning traps (near-synonyms, timeline confusions, etc.).  
These principles are embedded in the rules below.
</findings>

<general_rules>
1. Write the stem, four answer choices (“A”–“D”), and rationale **entirely in Kazakh**.  
2. Provide **exactly four** plausible answer choices. One and only one is correct.  
3. Each distractor must differ from the key by **one detail** (number, actor, date, term, etc.).  
4. Rotate the position of the correct answer so the keys distribute evenly across items.  
5. Avoid trivia so obscure that even specialists rarely know it; aim for benchmark-level difficulty.  
6. Do **not** include negative-logic stems unless unavoidable.  
7. Exclude giveaways such as “all of the above”, differing grammar, or absurd values.
</general_rules>

<challenge_tags>
Choose **at least one** tag for every item:  
‒ NEAR-SYNONYM ‒ NUMERIC ‒ ROLE ‒ TIMELINE ‒ POLYSEMY ‒ IDIOM ‒ HYPERNYM ‒ NEGATION
</challenge_tags>

<json_schema>

{{
  "question": "<Kazakh text>",
  "choices": {
    "A": "...",
    "B": "...",
    "C": "...",
    "D": "..."
  },
  "answer": "B",                      // capital letter only
  "tags": ["NEAR-SYNONYM","ROLE"],    // ≥1 tags
  "rationale": "Қысқа түсіндірме: неге B дұрыс, басқалары неге қате."
}}

Return **one JSON object per line** (JSONL). Do not wrap in Markdown.
\</json\_schema>

\<input\_fields>
• <topic> – a brief theme, e.g. “Наурыз мейрамы”
• <context> – background facts or excerpts for that topic (optional but recommended)
\</input\_fields>

\<examples\_placeholder>"""
f"""{examples}   <!--  Add your few-shot sample items here via an f-string -->
\</examples\_placeholder>

\<task\_prompt>
Using the <topic> and <context>, generate **N** new MCQs that satisfy every rule above.
Do not output explanations, headings, or any text outside the prescribed JSON objects.
\</task\_prompt>

""")
```

```python
from tqdm import tqdm
import re
import json

def process_batch(batch, batch_name):
    results = []
    max_try = 3
    
    for topic, context in tqdm(batch, desc=f"batch {batch_name}", total=10):
        # print(f"Processing topic: {topic}")
        # print("" * 20)
        # print(f"Processing context: {context}")

        p1 = f"""
        - **Topic**: {topic}
        - **Context**: {context}
        """.strip()

        current = 0
        while current < max_try:
            try:
                assistant_reply = genai.GenerativeModel("gemini-2.0-flash").generate_content(
                    contents=f"{GENERATE_PROMPT}\n\n<topic_and_context_for_generation>{p1}</topic_and_context_for_generation>.Output in JSONL format:\n\n"
                )
                json_match = re.search(r"\{.*\}", assistant_reply.text, re.DOTALL)
                if json_match:
                    res= json.loads(json_match.group())
                break
            except Exception:
                current += 1
                continue
        results.append({"topic": topic, "output": res if res else ""})
        
    return results
```

```python
import argparse, json, os, sys
from typing import Dict, Any

import pandas as pd
from datasets import load_dataset, Dataset
from dotenv import load_dotenv
import torch
from huggingface_hub.hf_api import HfFolder
from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.prompts import PromptTemplate
from tqdm import tqdm

load_dotenv()

_TEMPLATE = """
Сіз қазақ тілінде жауап беретін білімді, пайдалы көмекшісіз. Сұрақты және жауап нұсқаларын мұқият оқып,
ең дұрысын бір ғана әріппен (A, B, C, D) белгілеңіз.

Cұрақ: {prompt}
A) {a}
B) {b}
C) {c}
D) {d}

Жауап:""".strip()

prompt_4 = PromptTemplate(
    template=_TEMPLATE,
    input_variables=["prompt", "a", "b", "c", "d"],
)

def load_mc_dataset(raw_ds) -> Dataset:
    """Returns HF Dataset with .text, .answer, .category, .subcategory, .idx"""
    def build_text(ex: Dict[str, Any]) -> Dict[str, str]:
        return {
            "text": prompt_4.format(
                prompt=ex["question"],
                a=ex["A"],
                b=ex["B"],
                c=ex["C"],
                d=ex["D"],
            )
        }
    ds = raw_ds.map(build_text)
    #ds = ds.select(range(10))  # for testing
    print(ds["text"][0])
    #ds = ds.rename_column("formatted_answer", "answer")
    ds = ds.add_column("idx", [f"{i}" for i in range(len(ds))])
    return ds


def load_model_and_tokenizer(model_id: str, dtype: str = "bfloat16"):
    dtype_map = {
        "float16": torch.float16,
        "bfloat16": torch.bfloat16,
        "float32": torch.float32,
    }
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    tokenizer.pad_token = tokenizer.eos_token
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=dtype_map[dtype],
        trust_remote_code=True,
        device_map="auto",
    )
    model.eval()
    return model, tokenizer


def predict_option(model, tokenizer, prompt: str, apply_chat_template: bool = False) -> str:
    if apply_chat_template:
        print("Using chat template")
        messages = [{"role": "user", "content": [{"type": "text", "text": prompt}]}]
        prompt = tokenizer.apply_chat_template(messages, add_generation_prompt=True, tokenize=False)
    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {k: v.cuda() for k, v in inputs.items()}

    with torch.no_grad():
        logits = model(**inputs).logits[0, -1]

    # Four-choice logits (space-prefixed tokens)
    scores = {
        "A": logits[tokenizer(" A").input_ids[-1]],
        "B": logits[tokenizer(" B").input_ids[-1]],
        "C": logits[tokenizer(" C").input_ids[-1]],
        "D": logits[tokenizer(" D").input_ids[-1]],
    }
    return max(scores, key=scores.get)


def run_eval(ds: Dataset, model, tokenizer, apply_chat_template: bool = False) -> pd.DataFrame:
    rows = []
    #ds = ds.select(range(10))  # for testing
    for ex in tqdm(ds, total=len(ds)):
        pred = predict_option(model, tokenizer, ex["text"], apply_chat_template)
        rows.append(
            {
                "question": ex["question"],
                "A": ex["A"],
                "B": ex["B"],
                "C": ex["C"],
                "D": ex["D"],
                "idx": ex["idx"],
                "answer": ex["answer"].strip(),  # ensure no whitespace
                "predict": pred,

            }
        )
    df = pd.DataFrame(rows)
    df["correct"] = (df["answer"] == df["predict"]).astype(int)
    df.to_csv(f"predictions-{model.config._name_or_path.replace("/","-")}.csv", index=False)
    return df


```

```python
model, tok = load_model_and_tokenizer(model_id="meta-llama/Llama-3.3-70B-Instruct")
```

```python
ds = load_mc_dataset(raw_ds)
```

```python
acc = run_eval(ds, model, tok)
```
