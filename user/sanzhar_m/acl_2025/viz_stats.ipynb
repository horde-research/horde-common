```python
from pathlib import Path
import argparse
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

CAPTIONS = {
    "category_distribution.png":   "распределение по категориям",
    "subcategory_bars.png":        "распределение субкатегорий (цвет = категория)",
    "question_len_box.png":        "длина вопросов по категориям",
    "correct_pos_pie.png":         "позиция правильного ответа (pie)",
    "diacritics_radar.png":        "частоты казахских диакритик",
    "ttr_bar.png":                 "глобальный TTR по колонкам",
    "ttr_heatmap.png":             "TTR по категориям и колонкам",
}

def show_images(folder: Path) -> None:
    images = sorted(p for p in folder.glob("*.png"))
    if not images:
        print("No PNGs found in", folder)
        return

    for img_path in images:
        fig, ax = plt.subplots(figsize=(16, 9))
        ax.imshow(mpimg.imread(img_path))
        ax.axis("off")

        caption = CAPTIONS.get(img_path.name, img_path.stem.replace("_", " "))
        ax.set_title(f"{caption}  ({img_path.name})", fontsize=12)
        plt.tight_layout()
        plt.show()
```

```python
show_images(Path('viz/'))
```
